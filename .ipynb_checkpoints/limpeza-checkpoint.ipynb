```python
df
```

```python
#Alterando classificaçao de ocorrência
df.loc[df.ocorrencia_uf=='SP',['ocorrencia_classificacao']] = 'GRAVE'
```

```python
#Limpeza em coluna especifica
df.loc[df.ocorrencia_uf=='**',['ocorrencia_uf']] = pd.NA
```

```python
#Efetuando limpeza dos dados indesejados em tota tabela
df.replace(['**','***','###!','####','****','*****','NULL'], pd.NA,inplace=True)
```

```python
df.isna().sum()
```

```python
#Substindo valor dos <NA> por NI
df.fillna('NI',inplace=True)
```

```python
#Excluindo coluna, para efetivar colocar o inplace=True

df.drop(['ocorrencia_uf'],axis=1)
```

```python
#Elimina linhas com nulo
df.dropna()
#Elimina linhas nulas em uma coluna especifica
df.dropna(subset=['ocorrencia_uf'])
```

```python
#Elimina Duplicatas
df.drop_duplicates()
```
